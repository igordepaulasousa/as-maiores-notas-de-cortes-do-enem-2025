# ‚¨ÜDESCOBRINDO AS DEZ MAIORES NOTAS DE CORTE DO SISU **2025**

```python
# Agrupando os cursos e extraindo as suas notas m√°ximas  e classficando-as do maior para menor
top_notas = df.loc[df.groupby('NO_CURSO')['NU_NOTACORTE'].idxmax()]
top_notas = top_notas.sort_values(by='NU_NOTACORTE', ascending=False)
top_notas.head(10)
```

DESCOBRINDO AS 10 MAIORES NOTAS DE CORTE DO SISU 2025 NO ESP√çRITO SANTO

```python
#Filtrando apenas os cursos oferecidos no Esp√≠rito Santo
top_ES = (df['SG_UF_CAMPUS'] == 'ES')
#Agrupando os cursos e localizando as suas notas m√°ximas
top_ES = df.loc[df[top_ES].groupby('NO_CURSO')['NU_NOTACORTE'].idxmax()]
#Classificando as notas m√°ximas do amior para o menor
top_ES.sort_values(by='NU_NOTACORTE', ascending=False).head(10)
```
